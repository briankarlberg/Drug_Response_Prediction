```python
import pandas as pd
import time
from sklearn.svm import SVR
from sklearn.feature_selection import RFE
import argparse

# parser = argparse.ArgumentParser(
#     description = 'Drug index'
# )

# parser.add_argument(
#     'drug_frame_index',
#     type = int,
#     help = 'Input drug frame index number'
# )

# args = parser.parse_args()

dummy_loop = [0]

for i in dummy_loop:

    drug_frame_index = 0
    # drug_frame_index = args.drug_frame_index

    cell_line_file_read_start_time = time.time()

#     cell_lines_by_genes = pd.read_csv(
#         './data/CCLE_gene_expression_by_cell_line.tsv',
#         sep = '\t'
#     )

    print('Cell line file read time')
    print(
        round(
            time.time() - cell_line_file_read_start_time, 2
        )
    )
    print(' ')

    drug_resp_file_read_start_time = time.time()

    CTRPv2 = pd.read_csv(
        './data/drug_response_CTRPv2.tsv',
        sep = '\t'
    )

    print('Drug resp file read')
    print(
        round(
            time.time() - drug_resp_file_read_start_time, 2
        )
    )
    print(' ')

    CTRPv2_drug_sort = CTRPv2.sort_values('Drug')
    CTRPv2_drug_sort.reset_index(
        inplace = True,
        drop = True
    )

    CTRPv2_unique_compounds = CTRPv2_drug_sort.Drug.unique()
    
    compound = CTRPv2_unique_compounds[
        drug_frame_index
    ]

    drug_frame = CTRPv2[ 
        CTRPv2.Drug == CTRPv2_unique_compounds[
            drug_frame_index
        ]
    ]
    
    drug_frame.reset_index(
        inplace = True,
        drop = True
    )
    
    drug_cell_lines = drug_frame.Cell_line

    expression_cell_line_subset = cell_lines_by_genes[
        cell_lines_by_genes[
            'Cell_line'
        ].isin(
            drug_cell_lines
        )
    ]
    
#     del(cell_lines_by_genes)

    expression_cell_line_subset.reset_index(
        inplace = True,
        drop = True
    )

    gexp_tags = expression_cell_line_subset.iloc[:, :3]
    X_gexp = expression_cell_line_subset.iloc[ :, 3:]
#     X_gexp.reset_index(inplace = True, drop = True)

    X_gexp_cut = X_gexp.iloc[: , :100]
    
    X_gexp_cut_200 = X_gexp.iloc[: , :200]

    subset_cell_lines = expression_cell_line_subset.Cell_line
    
    subset_of_drug_frame_gexp_hits = drug_frame[
        drug_frame.Cell_line.isin(
            subset_cell_lines
        )
    ]

    y_response = subset_of_drug_frame_gexp_hits.Response
    
    estimator = SVR(kernel="linear")
    selector = RFE(estimator, n_features_to_select=10, step=1)

    start = time.time()
    break
    selector = selector.fit(X_gexp_cut, y_response)
    # selector = selector.fit(X_gexp_200, y_response)
    # selector = selector.fit(X_gexp, y_response)

    end = time.time() - start

    selected_features = X_gexp_cut.loc[
        :, selector.support_
    ]

    # selected_features = X_gexp_cut_200.loc[
    #     :, selector.support_
    # ]

    # selected_features = X_gexp.loc[
    #     :, selector.support_
    # ]

    selected_features.to_csv(
        'CTRP_features/selected_features/selected_features_'+compound+'_'+str(
            round(end, 2))+'.tsv',
        sep = '\t',
        index = False
    )
```
