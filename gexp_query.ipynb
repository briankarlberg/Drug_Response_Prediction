```python
def make_row(gene_expression_for_specific_cell_line):
    keys = []
    vals = []
    storDF_test = pd.DataFrame()
    for k, v in gene_expression_for_specific_cell_line['data']['values'].items():
        keys.append(k)
        vals.append(v)
    return(pd.DataFrame(vals, index = keys).T)
```

```python
start = time.time()

# for i in skip_range:
for i in skip_range:
    print(i) # keep
    cell_line_expression_block = G.query().V().hasLabel('Aliquot'
                        ).out('gene_expressions').skip(i).limit(
                        100).execute()
    Xn = pd.DataFrame()
    
#     gid_list = []
    label_list = []
    project_list = []
    cell_line_list = []
    
    for j in cell_line_expression_block:
#         print(j)
        cell_line = j['gid'].split(':')[2]
        project = j['gid'].split(':')[1]
        label = j['gid'].split(':')[0]
#         gid = j['gid']
        
        row_n = make_row(j) # j is gene_expression_for_single_cell_line
        Xn = pd.concat([Xn, row_n])
        
        cell_line_list.append(cell_line)
        project_list.append(project)
        label_list.append(label)
#         gid_list.append(gid)
        
    Xn.insert(0,'Label', label_list)
    Xn.insert(0,'Project', project_list)
    Xn.insert(0,'Cell_line', cell_line_list)
    
    Xn.to_csv('./expression_chunks/block_skip_'+str(i)+'.tsv',
                  sep = '\t', index = False)
    print(time.time() - start)
#     break
```
